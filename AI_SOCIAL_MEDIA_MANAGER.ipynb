```python
import pandas as pd
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="distilgpt2")
```

```python
data = {
    "post": ["AI meme","Study reel","Coding tips","Motivation","AI reel"],
    "likes": [120,350,220,180,500],
    "comments": [10,40,25,20,60],
    "time": ["18:00","20:00","19:00","17:00","21:00"],
    "type": ["meme","reel","image","image","reel"]
}

df = pd.DataFrame(data)
df
```

```python
print("üìä Audience Insight Report\n")

best_post = df.loc[df['likes'].idxmax()]
print("üî• Best Performing Post:", best_post['post'])

best_time = best_post['time']
print("‚è∞ Best Time to Post:", best_time)

best_type = df.groupby('type')['likes'].mean().idxmax()
print("üéØ Most Loved Content Type:", best_type)
```

```python
topic = best_type + " for students and tech audience"

prompt = f"Create an engaging instagram caption with emojis and hashtags about {topic}:"

result = generator(prompt, max_length=60, num_return_sequences=1)

caption = result[0]['generated_text']
print("üìù AI Generated Caption:\n")
print(caption)
```

```python
comment = "This post is awesome!"

prompt = f"Reply professionally to this comment: {comment}"

result = generator(prompt, max_length=50, num_return_sequences=1)

reply = result[0]['generated_text']
print("üí¨ AI Reply:\n")
print(reply)
```

```python
df['engagement'] = df['likes'] + df['comments']*2

viral = df.sort_values(by='engagement', ascending=False).head(1)

print("üöÄ Next Viral Content Suggestion:")
print(viral[['post','engagement']])
```
